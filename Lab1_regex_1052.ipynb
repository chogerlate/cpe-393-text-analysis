To split the Books we used this pattern
```
Rowling, J. K.

Harry Potter and the Chamber of Secrets / by J. K. Rowling.
```

````python
from logging import error
import os
import re

def preprocess_books(raw_text: str) -> str:
    """
    Reads and cleans up the raw Harry Potter text file.
    - Detects encoding automatically
    - Removes headers/footers, publisher info, copyright, dedications
    - Keeps only normal letters, digits, punctuation
    - Normalizes whitespace
    Returns:
        Cleaned text as a string
    """

    # --- Step 2: Normalize line endings ---
    text = raw_text.replace("\r\n", "\n").replace("\r", "\n")

    # --- Step 3: Remove Project Gutenberg headers/footers ---
    text = re.sub(r"\*\*\* START OF.*?\*\*\*", "", text, flags=re.DOTALL | re.IGNORECASE)
    text = re.sub(r"\*\*\* END OF.*?\*\*\*", "", text, flags=re.DOTALL | re.IGNORECASE)

    # --- Step 4: Remove publisher/copyright/metadata ---
    metadata_patterns = [
        r"copyright.*?\d{4}.*?\n",
        r"all rights reserved.*?\n",
        r"illustrations? by.*?\n",
        r"arthur a\. levine books.*?\n",
        r"scholastic press.*?\n",
        r"bloomsbury.*?\n",
        r"warner bros.*?\n",
        r"for .*\n",             # dedications
        r"in memory of.*?\n",
        r"isbn.*?\n",
    ]
    for pat in metadata_patterns:
        text = re.sub(pat, "", text, flags=re.IGNORECASE)

    # --- Step 5: Remove non-standard characters ---
    # Keep letters, numbers, common punctuation, whitespace
    text = re.sub(r"[^a-zA-Z0-9\s.,;:'\"!?()\-\n]", "", text)

    # --- Step 6: Collapse multiple spaces/newlines ---
    text = re.sub(r"[ \t]+", " ", text)         # collapse spaces
    text = re.sub(r"\n\s*\n\s*", "\n\n", text)  # keep paragraph breaks

    # --- Step 7: Strip leading/trailing whitespace ---
    return text.strip()



import os

def separate_harry_potter_books(text: str, output_dir: str):
    """
    Splits the concatenated Harry Potter text into separate book files.
    """

    # List of Harry Potter book titles in order
    book_titles = [
        "Harry Potter and the Philosopher's Stone",
        "Harry Potter and the Chamber of Secrets",
        "Harry Potter and the Prisoner of Azkaban",
        "Harry Potter and the Goblet of Fire",
        "Harry Potter and the Order of the Phoenix",
        "Harry Potter and the Half-Blood Prince",
        "Harry Potter and the Deathly Hallows"
    ]

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Directory '{output_dir}' created.")
    else:
        print(f"Directory '{output_dir}' already exists.")

    try:
        raw_text = text

        # Split the text by "Chapter 1\n" and remove the first empty element
        # (since the file starts with the delimiter)
        books_content = raw_text.split("Chapter 1\n")[1:]

        # Check if the number of books matches the number of titles
        if len(books_content) != len(book_titles):
            print("Warning: The number of books found does not match the number of titles.")
            print(f"Found {len(books_content)} books, but expected {len(book_titles)}.")

        # Save each book's content to a new file
        for i, content in enumerate(books_content):
            book_title = book_titles[i]
            # Add "Chapter 1" back to the start of each file
            full_content = "Chapter 1\n" + content
            file_name = f"{book_title.replace(' ', '_')}.txt"
            file_path = os.path.join(output_dir, file_name)

            with open(file_path, 'w', encoding='utf-8') as book_file:
                book_file.write(full_content)
                print(f"Successfully saved '{book_title}' to '{file_path}'")

        print("Separation complete! 🎉")

    except FileNotFoundError:
        print(f"Error: The file '{raw_text_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
raw_file_path = "/content/books.txt"
output_directory = "harry_potter_books"

# Read text file
with open(raw_file_path, 'r', encoding='utf-8') as file:
    raw_text = file.read()


cleaned_text = preprocess_books(raw_text)
separate_harry_potter_books(cleaned_text, output_directory)
````

````python
import os
import re

BOOK_DIR = "/content/harry_potter_books"

# regex for chapter headings
chapter_pattern = re.compile(r"^Chapter\s+(\d+)\s*\n", re.IGNORECASE | re.MULTILINE)

# iterate over book files
for filename in os.listdir(BOOK_DIR):
    if not filename.lower().endswith(".txt"):
        continue  # skip non-text files

    book_path = os.path.join(BOOK_DIR, filename)
    with open(book_path, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read().strip()

    # find all chapter starts
    matches = list(chapter_pattern.finditer(text))
    if not matches:
        print(f"No chapters found in {filename}")
        continue

    # create a folder for this book
    book_name = os.path.splitext(filename)[0]
    book_folder = os.path.join(BOOK_DIR, book_name)
    os.makedirs(book_folder, exist_ok=True)

    # split text into chapters
    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i + 1].start() if i + 1 < len(matches) else len(text)
        chapter_text = text[start:end].strip()

        chapter_title_line = chapter_text.splitlines()[0]
        chapter_num = re.search(r"\d+", chapter_title_line).group()
        chapter_filename = f"chapter_{chapter_num}.txt"

        with open(os.path.join(book_folder, chapter_filename), "w", encoding="utf-8") as out:
            out.write(chapter_text)

    print(f"✅ Split {filename} into {len(matches)} chapters")
````

````python
import os
import re
from glob import glob
from collections import Counter, defaultdict
import matplotlib.pyplot as plt

BOOK_DIR = "/content/harry_potter_books"

# Characters to count (with aliases)
CHARACTERS = {
    "Harry": ["Harry"],
    "Ron": ["Ron", "Weasley"],
    "Hermione": ["Hermione"],
    "Dobby": ["Dobby"],
    "Malfoy": ["Malfoy", "Draco"],
    "Voldemort": ["Voldemort", "You-Know-Who", "He-Who-Must-Not-Be-Named"],
    "Snape": ["Snape", "Severus"],
    "Dumbledore": ["Dumbledore", "Albus"],
    "Hagrid": ["Hagrid"],
    "Sirius": ["Sirius", "Black"],
}

def count_characters(text: str) -> Counter:
    """Count character mentions in given text."""
    counts = Counter()
    for char, aliases in CHARACTERS.items():
        for alias in aliases:
            counts[char] += len(re.findall(rf"\b{re.escape(alias)}\b", text))
    return counts

# --- Collect analysis ---
book_character_counts = {}
book_harry_dominance = {}

for book_folder in sorted(glob(os.path.join(BOOK_DIR, "*/"))):
    book_name = os.path.basename(os.path.normpath(book_folder))
    chapter_lengths = {}
    total_counts = Counter()
    harry_dom = []

    for chapter_file in sorted(glob(os.path.join(book_folder, "*.txt"))):
        chapter_name = os.path.splitext(os.path.basename(chapter_file))[0]

        with open(chapter_file, "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()

        # store chapter length
        chapter_lengths[chapter_name] = len(text)

        # count characters in chapter
        counts = count_characters(text)
        total_counts.update(counts)

        # find dominant character
        if counts:
            dominant = max(counts, key=counts.get)
            harry_dom.append((chapter_name, dominant))

    # save per book
    book_character_counts[book_name] = total_counts
    book_harry_dominance[book_name] = harry_dom


    print(f"\n📖 {book_name}")
    print(f"  Chapters: {len(chapter_lengths)}")

    # --- Harry dominance report ---
    print("  Harry dominance per chapter:")
    for chap, dom in harry_dom:
        print(f"    {chap}: {dom}")

    # --- Bar chart per book ---
    plt.figure(figsize=(10,6))
    names = list(total_counts.keys())
    values = [total_counts[name] for name in names]

    plt.bar(names, values)
    plt.title(f"Character Mentions in {book_name}")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()
````
